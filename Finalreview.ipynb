```python
import torch, torchaudio
import requests
import IPython.display as display
```

```python
hubert = torch.hub.load("Themozhi/hubert:main", "hubert_soft", trust_repo=True).cuda()
```

```python
acoustic = torch.hub.load("Themozhi/acoustic-model:main", "hubert_soft", trust_repo=True).cuda()
```

```python
hifigan = torch.hub.load("Themozhi/hifigan:main", "hifigan_hubert_soft", trust_repo=True).cuda()
```

```python
import torchaudio

# Provide the path of your file
file_path = r"C:\Users\beyon\Downloads\corrupted_audio (1).wav"
source, sr = torchaudio.load(file_path)
source = torchaudio.functional.resample(source, sr, 16000)
source = source.unsqueeze(0).cuda()
```

```python
with torch.inference_mode():
    # Extract speech units
    units = hubert.units(source)
    # Generate target spectrogram
    mel = acoustic.generate(units).transpose(1, 2)
    # Generate audio waveform
    target = hifigan(mel)
```

```python
from IPython.display import Audio

# Play the source and target audio
Audio(source.squeeze().cpu().numpy(), rate=16000)
```
